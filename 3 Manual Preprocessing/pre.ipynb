### Missing Value treatement
1. categorical (text) - mode
2. continuous (numerical) - mean, median
3. df.fillna(value)

```python
for i in X.columns:
    print(i)
```

```python
def replacer(df):
    cat = list(df.columns[df.dtypes == "object"])
    for i in df.columns:
        if i in cat:
            mode = df[i].mode()[0]
            df[i] = df[i].fillna(mode)
        else:
            mean = df[i].mean()
            df[i] = df[i].fillna(mean)
```
