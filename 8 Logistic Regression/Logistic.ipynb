# Logistic regression

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

### Step 1: Read the data

```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

```python
df["species"].value_counts()
```

### Step 2: Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep="first")
```

```python
df.shape
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.index
```

### Step 3 - Seperate X and Y (species)

```python
X = df.drop(columns=["species"])
Y = df[["species"]]
```

```python
X.head()
```

```python
Y.head()
```

### Apply preprocessing on X

```python
X.dtypes
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

```python
num_pipe
```

## Step 5 - Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, test_size=0.33, random_state=21
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6: Build logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

### Evaluate model

```python
# Training accuracy
model.score(xtrain, ytrain)
```

```python
# Test accuarcy
model.score(xtest, ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_estimator(model, xtrain, ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

```python
ypred_test = model.predict(xtest)
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

```python
(1 + 0.88 + 0.88)/3
```

```python
from sklearn.model_selection import cross_val_score
```

```python
f1_cv = cross_val_score(model, xtrain, ytrain, scoring="f1_macro", cv=5)
f1_cv
```

```python
f1_cv.mean()
```

### Step 8 - Predict out of sample data

```python
xnew = pd.read_csv("iris_sample.csv")
xnew
```

```python
num_pipe
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```

```python
preds = model.predict(xnew_pre)
probs = model.predict_proba(xnew_pre)
```

```python
preds
```

```python
model.classes_
```

```python
probs
```

```python
xnew["species_pred"] = preds
xnew[model.classes_] = probs
```

```python
xnew = xnew.round(4)
```
